```python
# Importing necessary libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print('Modules are imported !! ')
```

```python
# Importing Covid-19 dataset

corona_dataset_csv = pd.read_csv('covid19_Confirmed_dataset.csv')
corona_dataset_csv.head()
```

```python
# Always check the row and column (shape) of dataset

corona_dataset_csv.shape
```

```python
# Delete unwanted columns of data - In this case, lat and long is of no use

corona_dataset_csv.drop(['Lat', 'Long'], axis=1, inplace=True)
corona_dataset_csv.head(10)
```

```python
# Aggregating the rows by the country

corona_dataset_aggregated = corona_dataset_csv.groupby('Country/Region').sum()
corona_dataset_aggregated.head(20)
```

```python
corona_dataset_aggregated.shape
```

```python
# Visualize data related to a country (Eg: Compare China, Italy and Spain)

corona_dataset_aggregated.loc['China'].plot()
corona_dataset_aggregated.loc['Italy'].plot()
corona_dataset_aggregated.loc['Spain'].plot()
plt.legend()
```

```python
# View rate of change in virus number for few days

corona_dataset_aggregated.loc['China'][:3].plot()
```

```python
# Continue .. Finding first derivative of the curve

corona_dataset_aggregated.loc['China'].diff().plot()
```

```python
# Find maximum infection rate for China, Italy and Spain

print(corona_dataset_aggregated.loc['China'].diff().max())
print(corona_dataset_aggregated.loc['Italy'].diff().max())
print(corona_dataset_aggregated.loc['Spain'].diff().max())
```

```python
# Find maximum infection rate for all of the countries

countries = list(corona_dataset_aggregated.index)
max_infection_rate = []
for c in countries:
    max_infection_rate.append(corona_dataset_aggregated.loc[c].diff().max())
    
corona_dataset_aggregated['max_infection_rate'] = max_infection_rate
corona_dataset_aggregated.head()
```

```python
# Now we can create a new dataframe with only one needed column

corona_data = pd.DataFrame(corona_dataset_aggregated['max_infection_rate'])
corona_data.head()
```

```python
# Importing the world happiness dataset

happiness_report_csv = pd.read_csv('worldwide_happiness_report.csv')
happiness_report_csv.head()
```

```python
happiness_report_csv.shape
```

```python
# Remove useless columns

useless_cols = ['Overall rank', 'Score', 'Generosity', 'Perceptions of corruption']
happiness_report_csv.drop(useless_cols, axis = 1, inplace=True)
happiness_report_csv.head()
```

```python
# Changing the indices of the dataframe

happiness_report_csv.set_index('Country or region', inplace=True)
happiness_report_csv.head()
```

```python
# Join world happiness dataset with corona dataset

print(corona_data.shape)
print(happiness_report_csv.shape)
corona_data.head()
```

```python
data = corona_data.join(happiness_report_csv, how = 'inner')
data.head()
```

```python
# We need to see if there is any correlation between column related to the different happiness factors and the maximum infection
# rate in the countries

data.corr()
```

```python
# Visualization of the dataset

# Plotting max_infection_rate against GDP per capita

x = data['GDP per capita']
y = data['max_infection_rate']
sns.scatterplot(x, y)
```

```python
# Here we can see that x axis scaling is narrower than y axis, so we need to scale it

x = data['GDP per capita']
y = data['max_infection_rate']
sns.scatterplot(x, np.log(y))
```

```python
sns.regplot(x, np.log(y))
```

```python
# Plotting max_infection_rate against Social support

x = data['Social support']
y = data['max_infection_rate']
sns.scatterplot(x, np.log(y))
```

```python
sns.regplot(x, np.log(y))
```

```python
x = data['Healthy life expectancy']
y = data['max_infection_rate']
sns.scatterplot(x, np.log(y))
sns.regplot(x, np.log(y))
```

```python
x = data['Freedom to make life choices']
y = data['max_infection_rate']
sns.scatterplot(x, np.log(y))
sns.regplot(x, np.log(y))
```
